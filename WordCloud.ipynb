```python
from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import urllib
import requests
```

```python
text = ""
with open('sample.txt', encoding='utf-8') as f:
    text = ''.join(f.readlines())
```

```python
wc = WordCloud()
```

```python
wc.generate(text)
```

```python
wc.to_file('output1.png')
```

```python
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wc = WordCloud(max_font_size=40).generate(text)
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wc = WordCloud(background_color="white",max_font_size=40).generate(text)
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

### Masking the Cloud

```python
custom_mask = np.array(Image.open("krabby.png"))
wc = WordCloud(background_color="white", mask=custom_mask)
wc.generate(text)
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wc = WordCloud(background_color="white", mask=custom_mask, contour_width=3, contour_color='steelblue')
wc.generate(text)
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
wc.to_file('output2.png')
```

```python
custom_mask = np.array(Image.open("coloured-krabby.jpg"))
wc = WordCloud(background_color="white", mask=custom_mask)
wc.generate(text)
image_colors = ImageColorGenerator(custom_mask)
wc.recolor(color_func=image_colors)
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
wc.to_file('output3.png')
```

### Making wordcloud using Dictionary / String

```python
words = 'access guest guest apartment area area bathroom bed bed bed bed bed bedroom block coffee coffee coffee coffee entrance entry francisco free garden guest home house kettle kettle kitchen kitchen kitchen kitchen kitchen kitchen living located microwave neighborhood new park parking place privacy private queen room san separate separate shared space space space street suite time welcome'
```

```python
mask = np.array(Image.open(requests.get('http://www.clker.com/cliparts/O/i/x/Y/q/P/yellow-house-hi.png', stream=True).raw))
```

```python
#creating the function
def generate_wordcloud_white(words, mask):
    word_cloud = WordCloud(width = 512, height = 512, background_color='white', stopwords=STOPWORDS, mask=mask).generate(words)
    plt.figure(figsize=(10,8),facecolor = 'white', edgecolor='blue')
    plt.imshow(word_cloud)
    plt.axis('off')
    plt.tight_layout(pad=0)
    plt.show()
    
```

```python
generate_wordcloud_white(words, mask)
wc.to_file('output4.png')
```

```python
def generate_wordcloud_black(words, mask):
    word_cloud = WordCloud(width = 512, height = 512, background_color='black', stopwords=STOPWORDS, mask=mask).generate(words)
    plt.figure(figsize=(10,8),facecolor = 'white', edgecolor='blue')
    plt.imshow(word_cloud)
    plt.axis('off')
    plt.tight_layout(pad=0)
    plt.show()
masked = np.array(Image.open(requests.get('http://www.clker.com/cliparts/3/d/9/1/11949847661568287344heart_jon_phillips_01.svg.med.png', stream=True).raw))
generate_wordcloud_black(words, masked)
wc.to_file('output5.png')
```

```python
def generate_wordcloud_blue(words, mask):
    word_cloud = WordCloud(width = 512, height = 512, background_color='blue', stopwords=STOPWORDS, mask=mask).generate(words)
    plt.figure(figsize=(10,8),facecolor = 'white', edgecolor='blue')
    plt.imshow(word_cloud)
    plt.axis('off')
    plt.tight_layout(pad=0)
    plt.show()
img = np.array(Image.open(requests.get('http://www.clker.com/cliparts/3/9/5/4/1194985853388078539star_sergio_luiz_araujo__01.svg.med.png', stream=True).raw))
generate_wordcloud_blue(words, img)
wc.to_file('output6.png')
```
